```python
df_com = df_com.drop(['BLOCKID'], axis = 1)
```

```python
#missing values contain our dependent variable. for dependent variable can't fill with missing value which leads to 
#False information. hence dropping records which has missing values of dependent variable.

df_com.isnull().sum()['hc_mortgage_mean']  #There is 841 records with missing values of dependant variable
```

```python
df_com.dropna(subset = ['hc_mortgage_mean'], inplace = True)
```

```python
#Lets delete duplicate values if present

print('Before deleting duplicate values:',df_com.shape)

df_com.drop_duplicates(inplace = True)

print('After deleting duplicate values:',df_com.shape)
```

```python
col_check = df_com.isnull().sum().to_frame().reset_index()

col_check.head()
```

```python
col_check[col_check[0]>0]['index'].count()

#There are 39 columns with null values out of 80 columns after treating Blockid, hc_mortgage_mean columns and 
#duplicate values
```

```python
null_col = col_check[col_check[0]>0]['index'].tolist()

print(null_col,'\n\nCount of null Variables:',len(null_col))
```

```python
#If the column have less then or equal to 8 unique values then it will be considered as categorical column (fill missing
#values with mode) and if the column has greater than 8 unique values then it will be considered as continous (fill
#missing values with median)

for col in null_col:
    
    print(col,', Unique values:',df_com[col].nunique())
    
    if df_com[col].nunique()>8:
        
        df_com[col].fillna(df_com[col].median(), inplace = True)
        
    else:
        
        df_com[col].fillna(df_com[col].mode()[0],inplace = True)
        
#I see no mode value is filled since unique values is greater than 8 unique values of all variables.
```

```python
#cross check if there is null values in the dataset

df_com.isnull().sum() 

#Now the data is clear with missing values
```

```python
#Lets delete duplicates values of UID column because in train and test data may consists of duplicate UID column values

print('Shape before deleting duplicates of UID column:',df_com.shape)

df_com.drop_duplicates(subset = ['UID'], inplace = True)

print('Shape after deleting duplicates of UID column:',df_com.shape)
```

#### Data Modeling

```python
# Test if predicted variable is normally distributed

sns.distplot(df_com['hc_mortgage_mean'])

#predicted variable is slightly skewed.
```

```python
train = df_com[df_com['data_type'] == 'Train']
test = df_com[df_com['data_type'] == 'Test']
```

```python
a = ['UID','SUMLEVEL','state','state_ab','city','place','type','primary','data_type','pop_bins','hc_mortgage_mean']

#ignoring above variables which is some are categorical and some have no sense and has dependent variable

x_train = train.drop(a, axis = 1)
y_train = train['hc_mortgage_mean']

x_test = test.drop(a, axis = 1)
y_test = test['hc_mortgage_mean']
```

```python
from sklearn.preprocessing import MinMaxScaler

mn = MinMaxScaler()

x_train_mn = mn.fit_transform(x_train)
x_test_mn = mn.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
model = LinearRegression()

model.fit(x_train_mn,y_train)
```

```python
r2_score(y_test,model.predict(x_test_mn))
```

```python
r2_score(y_train,model.predict(x_train_mn)) #model performing not good lets go for VIF
```

```python
#Checking multicollearity

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate(df):
    vif = pd.DataFrame()
    vif['columns'] = df.columns
    vif['vif_value'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif

calculate(x_train)
```
